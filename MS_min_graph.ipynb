```python
from matplotlib import pyplot as plt
import numpy as np

x = np.arange(1,600)

y1 = -x+350
y2 = x ==125
y3 = -2*x+600

plt.xlim(0,600)
plt.ylim(0,600)
plt.plot(x,y1,'b',label='Total production')
plt.axvline(x=125, color='r',label = 'Demand for product A')
plt.plot(x,y3,'g', label='Processing time')
plt.legend(loc='upper right')
plt.xlabel('Gallons of product A')
plt.ylabel('Gallons of product B')
plt.title('Graphical Solution for the M&D Chemiclas problem')
plt.show()
```

```python
from matplotlib import pyplot as plt
import numpy as np

x = np.arange(1,600)

y1 = -x+350
y2 = x >=125
y3 = -2*x+600

plt.xlim(0,600)
plt.ylim(0,600)
plt.fill_between(x,y1,label='Total production',color ='green',alpha=0.2)
plt.axvline(x=125, color='r',label = 'Demand for product A', alpha = 0.3)
plt.fill_between(x,y3, label='Processing time', color = 'blue',alpha = 0.1)
plt.legend(loc='upper right')
plt.xlabel('Gallons of product A')
plt.ylabel('Gallons of product B')
plt.title('Graphical Solution for the M&D Chemiclas problem')
plt.show()
```

```python
from matplotlib import pyplot as plt
import numpy as np

x = np.arange(1,600)
#2x+3y

y1 = -x+350
y2 = x >=125
y3 = -2*x+600
y4 = -2/3*x+800/3 #optimal_object_value
y5 = -2/3*x+433.3 

plt.xlim(0,600)
plt.ylim(0,600)
plt.fill_between(x,y1,label='Total production',color ='green',alpha=0.2)
plt.axvline(x=125, color='r',label = 'Demand for product A', alpha = 0.3)
plt.fill_between(x,y3, label='Processing time', color = 'blue',alpha = 0.1)
plt.plot(x,y4, label = 'optimal_object_value')
plt.plot(x,y5, label = 'object_value_in feasiable region')
plt.legend(loc='upper right')
plt.xlabel('Gallons of product A')
plt.ylabel('Gallons of product B')
plt.title('Graphical Solution for the M&D Chemiclas problem')
plt.show()
```
