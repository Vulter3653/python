```python
from matplotlib import pyplot as plt
import numpy as np

x = np.arange(1,1400)

y1 = -7/10*x+630

plt.xlim(0,1400)
plt.ylim(0,1200)
plt.plot(x,y1,label='Cutting and Dyeing',color ='green')
plt.fill_between(x,y1,label='Cutting and Dyeing',color ='green', alpha = 0.1)
plt.xlabel('Number of Standard Bags')
plt.ylabel('Number of Deluxe Bags')
plt.title('Cutting and Dyeing Constraint')
plt.text(25, 640, '(0,630)')
plt.text(875, 50, '(900,0)')
plt.text(250, 250, '7/10S+1D=630\nCutting and Dyeing', rotation=-30)
plt.show()
```

```python
from matplotlib import pyplot as plt
import numpy as np

x = np.arange(1,1400)

y1 = -7/10*x+630
y2 = -3/5*x+720
y3 = -3/2*x+1062
y4 = -4/10*x+540


plt.xlim(0,1400)
plt.ylim(0,1200)
plt.plot(x,y1,color='green')
plt.fill_between(x,y1,label='Cutting and Dyeing',color ='green',alpha=0.1)
plt.plot(x,y2,color='red')
plt.fill_between(x,y2,label = 'Sewing', color='red', alpha= 0.1)
plt.plot(x,y3,color='blue')
plt.fill_between(x,y3, label='Finishing', color = 'blue',alpha = 0.1)
plt.plot(x,y4, color = 'orange')
plt.fill_between(x,y4, label = 'Inspection and Packaging', color ='orange',alpha = 0.1)
plt.legend(loc='upper right')
plt.xlabel('Number of Standard Bags')
plt.ylabel('Number of Deluxe Bags')
plt.title('Graphical Solution for the Par,Inc. problem')
plt.text(200, 200, 'Feasible\nRegion')
plt.show()
```
